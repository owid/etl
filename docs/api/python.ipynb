## Installing

The [`owid-catalog`](https://github.com/owid/etl/tree/master/lib/catalog
) provides API access to the data. Install it with:

```bash
$ pip install owid-catalog
```

## Searching for data

Firstly, import the `owid.catalog` module:

````python
from owid import catalog
````

You can search for an appropriate table of data using the `find(query)` method:

````python
catalog.find("covid")
````

````python
catalog.find("energy").head()
````

## Loading data

The `find()` method returns a data frame of results. Calling `.load()` on any row in the frame fetches the data into memory:

````python
results = catalog.find("energy")
df = results.iloc[2].load()
df.head()
````

In cases like the Covid-19 query above, where there was only one match, you can call `.load()` on the set of results directly:

````python
catalog.find("covid").load().head()
````

## Metadata

Data fetched from the API supports metadata at every level. Let's take a look by loading an energy mix dataset:

````python
df = catalog.find("energy_mix").iloc[0].load()
````

### Table metadata

Every data frame loaded is accompanied by a `.metadata` object that describes it in more detail.

````python
table_meta = df.metadata
table_meta.title
````

### Dataset metadata

Tables keep a reference to the dataset that they're part of, which also has information such as the sources of the data.

````python
dataset_meta = df.metadata.dataset
dataset_meta.sources[0].name
````

### Variable metadata

Each column in the data frame also has a `.metadata` object which describes that specific variable in additional detail.

````python
df.coal__pct_growth.metadata.title
````

````python
df.coal__pct_growth.metadata.unit
````

To learn more about what metadata fields are available, see the [`owid.catalog.meta`](https://github.com/owid/etl/tree/master/lib/catalog/owid/catalog/meta.py) module.

## Channels

Our catalog publishes multiple _channels_ of data, including:

- `garden`: curated data that has been harmonized and gone through Our World in Data's editorial process
- `meadow`: data from an upstream provider in unmodified form
- `backport`: data from our legacy data catalog
- `open_numbers`: a mirror of Gapminder's [Open Numbers](https://open-numbers.github.io) datasets

By default, `find()` will only search in `garden`, but you can also specify other channels to search in. This can be useful since our modern data catalog does not contain all the data used on our site.
